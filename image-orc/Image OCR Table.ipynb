```python
import layoutparser as lp

import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np
import cv2
```

```python
ocr_agent=lp.TesseractAgent.with_tesseract_executable(r'C:\Program Files\Tesseract-OCR\tesseract.exe',languages = ['eng'])
```
